```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
from statsmodels.stats.outliers_influence import variance_inflation_factor
import shap

# 设置绘图风格
plt.style.use('seaborn-v0_8')  # 或 'seaborn'
%matplotlib inline

file_path = r"C:\Users\DevanteLam\.cache\kagglehub\datasets\mahatiratusher\heart-disease-risk-prediction-dataset\versions\2\heart_disease_risk_dataset_earlymed.csv"
df = pd.read_csv(file_path)

# 查看数据前5行
print("数据预览：")
print(df.head())

# 检查数据基本信息
print("\n数据基本信息：")
print(df.info())
```

```python
# 计算相关性矩阵
corr = df.corr()

# 绘制热力图
plt.figure(figsize=(18, 15))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix (Original)")
plt.show()

# 设定阈值筛选高相关特征
threshold = 0.7
high_corr_pairs = []

# 遍历所有特征对
for i in range(len(corr.columns)):
    for j in range(i):
        if abs(corr.iloc[i, j]) > threshold:
            high_corr_pairs.append((corr.columns[i], corr.columns[j], corr.iloc[i, j]))

# 输出高相关特征对
print("高相关性特征对（相关性 > 0.7）：")
for pair in high_corr_pairs:
    print(f"{pair[0]} 和 {pair[1]}，相关性 = {pair[2]:.2f}")

# 手动移除高相关特征（示例：移除症状类特征）
features_to_drop = ["Shortness_of_Breath", "Fatigue", "High_Cholesterol"]
df_filtered = df.drop(columns=features_to_drop)

# 绘制筛选后的相关性矩阵
corr_filtered = df_filtered.corr()
plt.figure(figsize=(15, 12))
sns.heatmap(corr_filtered, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Filtered Correlation Matrix")
plt.show()
```

```python
# 提取特征和标签
X = df.drop(columns=["Heart_Risk"])
y = df["Heart_Risk"]

# 标准化数据
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 应用PCA并分析方差贡献
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
cumulative_variance = pca.explained_variance_ratio_.cumsum()

# 绘制累计方差图
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance by Principal Components")
plt.grid(True)
plt.show()

# 选择保留95%方差的主成分
n_components = (cumulative_variance <= 0.95).sum() + 1
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)
print(f"保留 {n_components} 个主成分，累计方差贡献率 = {cumulative_variance[n_components-1]:.3f}")

# 分析主成分权重
pca_weights = pd.DataFrame(
    pca.components_,
    columns=X.columns,
    index=[f"PC{i+1}" for i in range(n_components)]
)
print("\n主成分权重（绝对值前5特征）：")
for pc in pca_weights.index:
    sorted_weights = pca_weights.loc[pc].abs().sort_values(ascending=False)
    print(f"\n{pc}:")
    print(sorted_weights.head(5))
```

```python
# 定义三种数据集
# 原始数据
X_original = X
# 手动筛选后的数据
X_filtered = df_filtered.drop(columns=["Heart_Risk"])
# PCA降维后的数据
X_pca_transformed = X_pca

# 标准化数据（PCA已标准化，此处仅处理原始和筛选数据）
scaler_original = StandardScaler()
X_original_scaled = scaler_original.fit_transform(X_original)

scaler_filtered = StandardScaler()
X_filtered_scaled = scaler_filtered.fit_transform(X_filtered)

# 分割数据集
def split_data(X, y, test_size=0.2, random_state=42):
    return train_test_split(X, y, test_size=test_size, random_state=random_state)

# 原始数据
X_train_ori, X_test_ori, y_train_ori, y_test_ori = split_data(X_original_scaled, y)

# 筛选后的数据
X_train_fil, X_test_fil, y_train_fil, y_test_fil = split_data(X_filtered_scaled, y)

# PCA数据
X_train_pca, X_test_pca, y_train_pca, y_test_pca = split_data(X_pca_transformed, y)
```

```python
# 定义模型列表
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": GradientBoostingClassifier(n_estimators=100, random_state=42)
}

# 定义评估函数
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None
    
    # 计算指标
    accuracy = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba) if y_proba is not None else None
    report = classification_report(y_test, y_pred)
    
    # 绘制混淆矩阵
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {model.__class__.__name__}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    
    # 绘制ROC曲线（仅限概率模型）
    if y_proba is not None:
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        plt.figure(figsize=(6, 4))
        plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
        plt.plot([0, 1], [0, 1], linestyle='--')
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title(f"ROC Curve - {model.__class__.__name__}")
        plt.legend()
        plt.show()
    
    return accuracy, auc, report

# 在三种数据集上评估模型
datasets = {
    "Original Data": (X_train_ori, X_test_ori, y_train_ori, y_test_ori),
    "Filtered Data": (X_train_fil, X_test_fil, y_train_fil, y_test_fil),
    "PCA Data": (X_train_pca, X_test_pca, y_train_pca, y_test_pca)
}

results = {}

for data_name, (X_train, X_test, y_train, y_test) in datasets.items():
    print(f"\n=== 评估数据集: {data_name} ===")
    results[data_name] = {}
    for model_name, model in models.items():
        print(f"\n模型: {model_name}")
        accuracy, auc, report = evaluate_model(model, X_train, X_test, y_train, y_test)
        results[data_name][model_name] = {
            "Accuracy": accuracy,
            "AUC": auc,
            "Report": report
        }
        print(f"准确率: {accuracy:.4f}")
        print(f"AUC: {auc:.4f}" if auc else "AUC: Not Available")
        print("分类报告:\n", report)
```

```python
import numpy as np
import shap
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# 训练模型（使用PCA后的数据）
model_rf_pca = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf_pca.fit(X_train_pca, y_train_pca)

# 创建SHAP解释器
explainer = shap.TreeExplainer(model_rf_pca)

# 计算SHAP值（针对类别1）
shap_values = explainer.shap_values(X_test_pca)

# 验证SHAP值类型和维度
print("SHAP值类型:", type(shap_values))
print("SHAP值维度:", np.array(shap_values).shape)

# 提取类别1（高风险）的 SHAP 值
shap_values_class1 = shap_values[:, :, 1]  # 提取所有样本和特征，针对类别1

# 验证维度
print("SHAP值维度（类别1）:", shap_values_class1.shape)
print("输入数据维度:", X_test_pca.shape)

# 绘制SHAP摘要图
plt.figure()
shap.summary_plot(
    shap_values_class1,
    X_test_pca,
    feature_names=[f"PC{i+1}" for i in range(X_test_pca.shape[1])],
    plot_type="bar"
)
plt.title("SHAP Feature Importance (PCA Data)")
plt.show()
```
